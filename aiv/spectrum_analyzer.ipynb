```python
import matplotlib.pyplot as plt
import csv
import datetime
import matplotlib.dates as mdates
import os
import numpy as np
import pandas as pd
```

```python
pklpath = "/home/s/stalder/sa/"
pklfile = 'Bias3K_S11_stats.pkl'
df = pd.read_pickle(pklpath+pklfile)

for seg,array in df.items(): 
    mjd=[]
    mean=[]
    stdev=[]
    for i in range(0,len(array)):
        mjd.append(array[i][0])
        mean.append(array[i][1])
        stdev.append(array[i][2])

    plt.plot(mjd,mean, label='From TS7')
    ax = plt.gca()
    #plt.xscale("log")
    #plt.yscale("log")
    #plt.xlim([4e5,3e7])
    #plt.ylim([2.5e4,2.6e4])
    plt.xlabel('MJD')
    plt.ylabel('mean (ADU)')
    plt.title('STDEV: '+pklfile+' seg:'+seg)
    plt.show()

    plt.plot(mjd,stdev, label='From TS7 amp:'+seg)
    ax = plt.gca()
    #plt.xscale("log")
    #plt.yscale("log")
    #plt.xlim([4e5,3e7])
    #plt.ylim([0,5])
    plt.xlabel('MJD')
    plt.ylabel('STDEV (ADU)')
    plt.title('STDEV: '+pklfile+' seg:'+seg)
    plt.show()
```

```python
from dateutil import parser
from astropy.time import Time

# Get the list of all files and directories
path = "/home/s/stalder/sa/sa_data_20221216/"
dir_list = os.listdir(path)
dir_list.sort()
#print(dir_list)

fileindex=0


for file in dir_list:
    freq = []
    trace = []
    line = 0
    date = Time(parser.parse(os.path.splitext(file)[0]))  
    mjd = date.mjd

#plot individual scans    
    # with open(path+file,'r') as csvfile:
    #     spectrum_meas = csv.reader(csvfile, delimiter=',')
    #     for row in spectrum_meas:
    #         if line > 1:
    #             freq.append(float(row[0]))
    #             trace.append(float(row[1]))
    #         line=line+1
    #     plt.plot(freq,trace, label='From SA')
    #     ax = plt.gca()
    #     plt.xlabel('Freq (Hz)')
    #     plt.ylabel('dBm')
    #     plt.title('SA Output: '+file)
    #     plt.show()

            
#plot averages
    averagetime=60. # in seconds
    with open(path+file,'r') as csvfile:
        spectrum_meas = csv.reader(csvfile, delimiter=',')
        for row in spectrum_meas:
            if line > 1:
                freq.append(float(row[0]))
                trace.append(float(row[1]))
            line=line+1
        freqarr=np.asarray(freq)
        tracearr=np.asarray(trace)
        if fileindex == 0:
            freqavg=freqarr
            traceavg=tracearr/averagetime  
        else:
            traceavg=traceavg+tracearr/averagetime

        fileindex=fileindex+1
        if (fileindex % averagetime) == 0:    
            traceavg=traceavg
#            plt.plot(freqavg,traceavg, label='From SA')
            plt.plot(freqavg,10**(traceavg/10)/freqavg, label='From SA')
            ax = plt.gca()
            plt.xscale("log")
            plt.yscale("log")
            plt.xlim([4e5,3e7])
#            plt.xlim([1e9,3e9])
            plt.xlabel('Freq (Hz)')
            plt.ylabel('Power (mW / Hz)')
            plt.title('SA Scan: '+str(mjd))
            plt.show()
            fileindex=0
```
